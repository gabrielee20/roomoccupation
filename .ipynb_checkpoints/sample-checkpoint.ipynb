# Occupazione della stanza

Per prima cosa iniziamo a recuperare il dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

X = pd.read_csv('./data/datatraining.txt')
print(X.head())
print(X.columns)


X['date'] = pd.to_datetime(X['date'], format='mixed')
X['weekday'] = X['date'].dt.day_name()

plt.figure(figsize=(10, 6))
sns.stripplot(data=X, x='weekday', y='CO2', jitter=True)
plt.xticks(rotation=45)
plt.title('Umidità per giorno della settimana')
plt.xlabel('Giorno della settimana')
plt.ylabel('Umidità')
plt.show()

# Calcolo conteggi
counts = X.groupby('weekday')['Occupancy'].sum()

sns.barplot(x=counts.index, y=counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.xlabel('Giorno della settimana')
plt.ylabel('Numero di presenti')
plt.title('Presenze totali per giorno della settimana')
plt.show()
```
